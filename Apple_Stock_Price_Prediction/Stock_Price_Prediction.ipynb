```python
# Select features (X) and target variable (y)
# We will use Open, High, Low, and Volume as features to predict the Adjusted Close price

features = ['Open', 'High', 'Low', 'Volume']
target = 'Adj Close'

X = stock_data[[(feature, 'AAPL') for feature in features]]
y = stock_data[(target, 'AAPL')]

# Display the first few rows of features and target
print("Features (X):")
display(X.head())
print("\nTarget (y):")
display(y.head())
```

```python
from sklearn.model_selection import train_test_split

# Determine the split point (e.g., 80% for training, 20% for testing)
train_size = int(len(stock_data) * 0.8)

# Split the data chronologically
X_train, X_test = X[0:train_size], X[train_size:len(stock_data)]
y_train, y_test = y[0:train_size], y[train_size:len(stock_data)]

# Display the shapes of the resulting sets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# Create new features, including technical indicators
def create_features(df):
    df = df.copy()
    df['year'] = df.index.year
    df['month'] = df.index.month
    df['day'] = df.index.day
    df['dayofweek'] = df.index.dayofweek
    df['dayofyear'] = df.index.dayofyear
    df['weekofyear'] = df.index.isocalendar().week
    df['quarter'] = df.index.quarter
    df['is_month_start'] = df.index.is_month_start.astype(int)
    df['is_month_end'] = df.index.is_month_end.astype(int)
    df['is_quarter_start'] = df.index.is_quarter_start.astype(int)
    df['is_quarter_end'] = df.index.is_quarter_end.astype(int)
    df['is_year_start'] = df.index.is_year_start.astype(int)
    df['is_year_end'] = df.index.is_year_end.astype(int)

    # Using the correct column names from the MultiIndex for calculations
    adj_close = df[('Adj Close', 'AAPL')]

    # Add Rolling Mean and Lag features
    df[('Rolling_Mean', 'AAPL')] = adj_close.rolling(window=10).mean()
    df[('Lag_1', 'AAPL')] = adj_close.shift(1)

    # Calculate Technical Indicators
    # MACD (Moving Average Convergence Divergence)
    # Typically uses 12-day EMA, 26-day EMA, and 9-day EMA for the signal line
    ema_12 = adj_close.ewm(span=12, adjust=False).mean()
    ema_26 = adj_close.ewm(span=26, adjust=False).mean()
    df[('MACD', 'AAPL')] = ema_12 - ema_26
    df[('Signal_Line', 'AAPL')] = df[('MACD', 'AAPL')].ewm(span=9, adjust=False).mean()

    # RSI (Relative Strength Index)
    # Typically uses a 14-period setting
    delta = adj_close.diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    avg_gain = gain.ewm(span=14, adjust=False).mean()
    avg_loss = loss.ewm(span=14, adjust=False).mean()
    rs = avg_gain / avg_loss
    df[('RSI', 'AAPL')] = 100 - (100 / (1 + rs))

    # EMA (Exponential Moving Average)
    # Adding a couple of common EMA periods
    df[('EMA_50', 'AAPL')] = adj_close.ewm(span=50, adjust=False).mean()
    df[('EMA_200', 'AAPL')] = adj_close.ewm(span=200, adjust=False).mean()


    return df

stock_data_features = create_features(stock_data)

# Drop rows with NaN values created by rolling mean, lag, and technical indicators
stock_data_features.dropna(inplace=True)

# Display the first few rows with new features
display(stock_data_features.head())
```

```python
# Print all column names and their types in stock_data_features
print("Columns and types of stock_data_features:")
for col, dtype in stock_data_features.dtypes.items():
    print(f"Column: {col}, Type: {dtype}")
```

```python
from sklearn.model_selection import train_test_split

# Determine the split point (e.g., 80% for training, 20% for testing)
# Use the length of the dataframe with new features
train_size = int(len(stock_data_features) * 0.8)

# Split the data chronologically
X_train, X_test = X[0:train_size], X[train_size:len(stock_data_features)]
y_train, y_test = y[0:train_size], y[train_size:len(stock_data_features)]

# Display the shapes of the resulting sets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# Select features (X) and target variable (y) again with new features
target = 'Adj Close'

# Based on the output of stock_data_features.columns and the dtypes,
# the columns have a mixed structure.
# Original yfinance columns have a MultiIndex like ('Adj Close', 'AAPL'), ('Close', 'AAPL'), etc.
# New features created in create_features like 'year', 'month', etc. have a single level index.
# 'Rolling_Mean', 'Lag_1', and the new technical indicators were added with a MultiIndex.

# Define the list of columns to select for features (X), handling the mixed index
# Create a list of all feature column names exactly as they appear in stock_data_features.columns
X_columns = [col for col in stock_data_features.columns if col != (target, 'AAPL')]


# Select the features and target from the dataframe with new features
X = stock_data_features[X_columns]
y = stock_data_features[(target, 'AAPL')]


# Display the first few rows of features and target
print("Features (X) with new features:")
display(X.head())
print("\nTarget (y):")
display(y.head())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Initialize different models
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest Regressor": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting Regressor": GradientBoostingRegressor(n_estimators=100, random_state=42)
}

# Dictionary to store model performance
performance_metrics = {}

# Train and evaluate each model
for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Evaluate the model
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    performance_metrics[name] = {
        "MAE": mae,
        "MSE": mse,
        "RMSE": rmse,
        "R-squared": r2,
        "R-squared-train": r2_score(y_train, model.predict(X_train))
    }
    print(f"{name} Evaluation:")
    print(f"  MAE: {mae:.4f}")
    print(f"  MSE: {mse:.4f}")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  R-squared: {r2:.4f}")
    print(f"  R-squared (train): {r2_score(y_train, model.predict(X_train)):.4f}")
    print("-" * 30)

# Display the performance metrics for all models
print("\nPerformance Metrics for all Models:")
display(pd.DataFrame(performance_metrics).T)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Define the parameter grid for GridSearchCV
# We will tune a few key hyperparameters for demonstration
param_grid = {
    'n_estimators': [50, 100, 200], # Number of trees in the forest
    'max_depth': [None, 10, 20],   # Maximum depth of the tree
    'min_samples_split': [2, 5, 10], # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4]    # Minimum number of samples required to be at a leaf node
}

# Initialize the RandomForestRegressor model
rf = RandomForestRegressor(random_state=42)

# Initialize GridSearchCV
# cv=5 means 5-fold cross-validation
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

# Perform Grid Search on the training data
print("Performing Grid Search for RandomForestRegressor...")
grid_search.fit(X_train, y_train)

# Get the best hyperparameters and the best model
best_params = grid_search.best_params_
best_rf_model = grid_search.best_estimator_

print("\nBest hyperparameters found by Grid Search:")
print(best_params)

# Evaluate the best model on the test data
y_pred_grid = best_rf_model.predict(X_test)

# Evaluate the best model
mae_grid = mean_absolute_error(y_test, y_pred_grid)
mse_grid = mean_squared_error(y_test, y_pred_grid)
rmse_grid = np.sqrt(mse_grid)
r2_grid = r2_score(y_test, y_pred_grid)

print("\nEvaluation of the best RandomForestRegressor found by Grid Search:")
print(f"  MAE: {mae_grid:.4f}")
print(f"  MSE: {mse_grid:.4f}")
print(f"  RMSE: {rmse_grid:.4f}")
print(f"  R-squared: {r2_grid:.4f}")
```
